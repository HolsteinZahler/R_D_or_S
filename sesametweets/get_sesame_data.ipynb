```python
# Import the Twython class
import math as math
from twython import Twython
import pandas as pd
import json
# Loading credentials
from twitter_config import api_key,s_api_key
```

```python
# Instantiate an object
twitter = Twython(api_key, s_api_key)
```

```python
# Create our query
query = {'q': 'learn python',
        'result_type': 'popular',
        'count': 10,
        'lang': 'en',
        }
```

```python


# Search tweets
dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': []}
for status in twitter.search(**query)['statuses']:
    dict_['user'].append(status['user']['screen_name'])
    dict_['date'].append(status['created_at'])
    dict_['text'].append(status['text'])
    dict_['favorite_count'].append(status['favorite_count'])

# Structure data in a pandas DataFrame for easier manipulation
df = pd.DataFrame(dict_)
df.sort_values(by='favorite_count', inplace=True, ascending=False)
df.head(5)
```

```python
user_names = ['AbbyCadabbySST',
'SesameErnie',
'bertsesame',
'Grover',
'BigBird',
'OscarTheGrouch',
'MeCookieMonster']
```

```python
num_tweets_to_fetch_per_char = 1000
num_queries = math.ceil(num_tweets_to_fetch_per_char/200)
tweet_idx = 0
sesame_tweets_df = pd.DataFrame(columns=['user_name','twitter_id','created_at','text'])

for user_name in user_names:
    tweets_left = num_tweets_to_fetch_per_char
    prev_max_id=0
    tweet_ct = 0
    for j in range(0,num_queries):
        if tweets_left < 200:
            querry_ct = tweets_left
        else:
            query_ct = 200
        if j==0:
            user_tweets = twitter.get_user_timeline(screen_name=user_name, include_rts=True, count=query_ct)
        else:
            user_tweets = twitter.get_user_timeline(screen_name=user_name, include_rts=True, count=query_ct, max_id=prev_max_id)
        for tweet in user_tweets:
            #tweet['text'] = Twython.html_for_tweet(tweet)
            if (tweet['text'][0:2]!='RT') and (tweet['text']!='Poof!'):
                #print(tweet_ct,tweet['id'],tweet['user']['id'],user_name,tweet['created_at'],tweet['text'])
                temp_vect = [user_name, tweet['user']['id'], tweet['created_at'][4:11]+tweet['created_at'][26:31],tweet['text']]
                sesame_tweets_df.loc[tweet_idx]=temp_vect
                tweet_ct+=1
                tweet_idx+=1
        prev_max_id=user_tweets[-1]['id']
    print(user_name,tweet_ct)

sesame_tweets_df
```

```python
def date_conversion(date_string):
    months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    not_done = True
    j = 0
    while not_done:
        if date_string[0:3]==months[j]:
            month_idx = j+1
            not_done = False
        elif j<11:
            j+=1
        else:
            return 'NaN'
            not_done = False
    return str(month_idx)+'/'+date_string[4:6]+'/'+date_string[7:11]
        
```

```python
date_conversion('Jan 19 1980')
```

```python
sesame_tweets_df['created_at'] = sesame_tweets_df['created_at'].apply(date_conversion)
```
